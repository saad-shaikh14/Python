Use Pandas ``groupby()`` and ``pivot_table()`` methods to construct 8 different summary statistics. They must include the following Pandas techniques:

`````python
# write your code here

#groupby describe()

DF1.groupby('Countries').describe()    #descriptive statistics which describes the dataframe
                                       #count, mean, standard deviation, min, max, 25th, 
                                       #50th, and 75th percentile
`````

`````python
#.describe()

DF1[["Countries","Literacy rate, youth male"]].describe()
`````

`````python
#groupby aggregation()

DF1.groupby('Years').aggregate(['min', 'median', 'max'])    #groupby with years, the data aggregations of mathematical
                                                            #expression such as mean, medain, and mode.
                                                            #aggregate basically collects two or more 
                                                            #values and combine them which is in this case are combined
                                                            #based on the groupby values i.e 'Years'
`````

`````python
#groupby filtering()

def filter_func(x):
    return x['Literacy rate, youth male'].std() < 10     #defined a function for filter which will check the standard  
                                                         #deviation on the corresponding column and if the std is           
                                                         #less than 10 it will drop the rows which has std greater than 10
                                                         
DF1.groupby('Countries').filter(filter_func)             #using groupby with countries to apply filtering on the dataframe 
`````

`````python
#groupby transform()

DF1.groupby('Years').transform(lambda x: x - x.mean())   #.transform is used to transform the entire dataframe not 
                                                         #just specific columns. the lambda function goes row by 
                                                         #row and takes the mean of rows for the entire dataframe           
`````

`````python
#groupby apply()                       #the apply method allows you to apply arbitary function to the group
                                       #it takes a dataframe and returns either a pandas dataframe or a scalar         

def norm_by_Countries(x):              #normalizing by the columns countries
    x['Unemployment, youth male'] /= x['Total Population, male'].sum() #dividing the values of column 
    return x

DF1.groupby('Countries').apply(norm_by_Countries)
`````

`````python
#groupby using pivot table

DF1.groupby('Years')[['Total Population, male']].mean()  #aggregating total population by years using pivot table
`````

`````python
#groupby using pivot table

DF1.pivot_table('Literacy rate, youth male', index='Years', columns='Countries', margins=True) #aggregating literacy rate 
                                                                                               #among youth by using index as 
                                                                                               #years and columns as countries
`````

`````python
#groupby using country_groups  

DF1.droplevel(level = 1).groupby(country_groups).mean()['Total Population, male']
`````
